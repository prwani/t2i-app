# Prompt improvement with a text LLM

This notebook uses the app's prompt-improvement service, which calls the configured Azure OpenAI text model and returns a stronger image prompt. Set `RUN_PROMPT_IMPROVEMENT=1` to execute the LLM call.

```python
from pathlib import Path
import os
import sys

from dotenv import load_dotenv

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "packages" / "t2i_core").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.insert(0, str(PROJECT_ROOT))
sys.path.insert(0, str(PROJECT_ROOT / "packages" / "t2i_core" / "src"))
load_dotenv(PROJECT_ROOT / ".env")

RUN_PROMPT_IMPROVEMENT = os.getenv("RUN_PROMPT_IMPROVEMENT") == "1"
print("Paid LLM call enabled:", RUN_PROMPT_IMPROVEMENT)
```

```python
from app.services import improve_prompt_with_ai
from t2i_core import Settings

settings = Settings()
print("Prompt-improvement model:", settings.eval_model_config.judge_deployment)
```

```python
raw_prompt = "a mug on a table, ad photo"
scenario = "text-to-image ecommerce product hero"

if RUN_PROMPT_IMPROVEMENT:
    enhanced_prompt = await improve_prompt_with_ai(raw_prompt, scenario)
else:
    enhanced_prompt = (
        "Photorealistic ecommerce product hero image of a matte ceramic mug on a natural oak table, "
        "warm morning side light, soft realistic shadow, clean premium composition, shallow depth of field, "
        "neutral background, no text, no logos, no extra objects."
    )
    print("Set RUN_PROMPT_IMPROVEMENT=1 to replace this local example with an Azure OpenAI response.")

print("Original prompt:\n", raw_prompt)
print("\nEnhanced prompt:\n", enhanced_prompt)
```

Use the enhanced prompt directly with `MAIImageProvider.generate(...)` or `GPTImageProvider.generate(...)`. For edits, include explicit preservation instructions such as "preserve the source product shape, label, and lighting".
